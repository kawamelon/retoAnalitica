```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_c=pd.read_csv("covid19_tweets.csv")
data_c
```

```python
data_c.shape
```

# **Nueva tabla para hacer el analisis**

```python
new_data = data_c[['user_name','user_followers','user_location','user_verified']]
new_data
```

# **Tabla de orden descendente de followers**

```python
new_data = new_data.sort_values(by='user_followers', ascending=False)
new_data
```

# **Tabla para eliminar repeticiones de user_name y dejar el que tenga mayor numero de followers**

```python
new_data = new_data.drop_duplicates(subset='user_name', keep='first')
new_data
```

# **Tabla Influencers**
*10.000 a 100.000 seguidores*

```python
inf_10mil = []
inf_10mil = (new_data[new_data["user_followers"] < 99999])
inf_10mil = (inf_10mil[inf_10mil["user_followers"] > 9999])
inf_10mil.sort_values("user_followers", ascending=False)
```

Analisis estadisitico

```python
media = np.mean(inf_10mil['user_followers'])
print('Media aritmetica:', media)

mediana = np.median(inf_10mil['user_followers'])
print("Mediana:", mediana)

desv = np.std(inf_10mil['user_followers'])
print("desviacion estandar:", desv)
print()
sns.boxplot(inf_10mil, showfliers=False)
```

En este rango, no podemos ver mucha diferencia entre las primeras cuentas con m√°s seguidores. De igual manera, tambi√©n podemos observar como el tipo de cuentas que se encuentran aqu√≠, son cuentas personales.

# **LOCACIONES**

```python
cont_loca_inf = inf_10mil['user_location'].value_counts()

data_loca_inf = pd.DataFrame({'Location': cont_loca_inf.index, 'Frequency': cont_loca_inf.values})

print(data_loca_inf.head(20))

plt.figure(figsize=(12, 6))
plt.bar(data_loca_inf['Location'][:20], data_loca_inf['Frequency'][:20])
plt.xlabel('Ubicaci√≥n de Influencers')
plt.ylabel('Frecuencia')
plt.title('Top 20 Ubicaciones de Influencers')
plt.xticks(rotation=90)
plt.show()
```
