```python
import pandas as pd
import numpy as np
import date as dt
```

```python
df = pd.read_csv('covid19_tweets.csv')
df
```

```python
col1 = df["user_followers"]
col2 = df["user_friends"]
col3 = df["user_favourites"]

de1 = np.std(col1)
de2 = np.std(col2)
de3 = np.std(col3)

print("La desviaci√≥n est√°ndar de la lista user_followers es: ", de1)
print("\nLa desviaci√≥n est√°ndar de la lista user_friends es: ", de2)
print("\nLa desviaci√≥n est√°ndar de la lista user_favourites es: ", de3)
```

```python
median1 = col1.median(axis=0)
median2 = col2.median(axis=0)
median3 = col3.median(axis=0)

print("La mediana de la lista user_followers es: ", median1)
print("\nLa mediana de la lista user_friends es: ", median2)
print("\nLa mediana de la lista user_favourites es: ", median3)
```

```python
prom1 = col1.mean()
prom2 = col2.mean()
prom3 = col3.mean()

print("La media de la lista user_followers es: ", prom1)
print("\nLa media de la lista user_friends es: ", prom2)
print("\nLa media de la lista user_favourites es: ", prom3)
```
